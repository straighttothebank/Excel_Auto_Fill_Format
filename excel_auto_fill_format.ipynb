```python
import openpyxl as op
import os
import pandas as pd
```

```python
#Carregando os dados
base = op.load_workbook('Dados.xlsx')
inputs = op.load_workbook('Entrada.xlsx')
df_base = pd.read_excel('Dados.xlsx')
df_inputs = pd.read_excel('Entrada.xlsx')
```

```python
#Setando as worksheets
base_ws = base['Sheet1']
inputs_ws = inputs['Sheet1']
```

```python
#Checking if inputs columns are repeating
same_columns = 0
for i in df_inputs.columns:
    for j in df_base.columns:
        if i == j:
            same_columns += 1
print("The number of columns are {} and {} of them are equal to base columns \n".format(same_columns,len(df_base.columns)))

#Checking for null values
print("List of columns with null values:{} \n".format(df_inputs.columns[df_inputs.isna().any()].tolist()))
print("Number of columns with null values is {} \n".format(len(df_inputs.columns[df_inputs.isna().any()])))

#Verificar se apresentam valores acima do permitido (>1000)
if True in df_inputs.values>1000:
    print("There is values above 1000")
else:
    print("Don't have any value above 1000")
```

```python
#Setting the number of rows and columns
last_row = df_base.shape[0]
n_columns = df_base.shape[1]
l = last_row
last_row , n_columns
```

```python
#Getting the columns names
columns_names_base = df_base.columns
columns_names_input = df_inputs.columns
```

```python
m=0
while m < n_columns:
    for i in df_inputs.columns:
        #verificar coluna igual
        for j in df_base.columns:
            if i == j:
                #pegar a quantidade de linhas que possui a Entrada na coluna i
                n_rows_input = df_inputs[i].shape[0]
                #pegar os index das colunas
                cb = columns_names_base.get_loc(j)
                ci = columns_names_input.get_loc(j)
                end_row = last_row + n_rows_input
                #iniciando na ultima linha, os dados vao ser inseridos 1 por 1
                for k in range(last_row, end_row):
                    #pegar a celula de refencia da base_ws (ultima linha, coluna j)
                    r = int(k + 2)
                    cellref = base_ws.cell(row=r, column=cb+1)
                    #inserir o valor na celula de posicao (k,j) pegando a celula de entrada na posicao (p,c)
                    cellref.value = df_inputs.iloc[k - l, ci]
                m = m + 1
m
```
